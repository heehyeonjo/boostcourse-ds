# 주피터 노트북 사용법

* Shift + Enter 키를 누르면 셀이 실행되고 커서가 다음셀로 이동
* ESC 키를 누르면 다시 편집 상태로 돌아옴
* ESC 키를 누르고
    * a 키를 누르면 위에 셀이 추가됨
    * b 키를 누르면 아래에 셀이 추가됨
    * dd 키를 누르면 셀이 삭제됨
    * m 키를 누르면 문서 셀로 변경됨
    * y 키를 누르면 코드 셀로 변경됨
